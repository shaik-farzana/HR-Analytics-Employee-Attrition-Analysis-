```python
import re
df.columns = [re.sub(r'(?<!^)(?=[A-Z])', '_', col).lower() 
              for col in df.columns]
```

```python
df.columns
```

```python
df=df.rename(columns={'emp_i_d':'emp_id'})
```

```python
df['years_with_curr_manager']=df['years_with_curr_manager'].fillna(df['years_with_curr_manager'].median())
```

```python
df.drop(['monthly_rate'],axis=1,inplace=True)
```

```python
df.drop(['employee_count'],axis=1,inplace=True)
```

```python
df.columns
```

```python
df['experience_level']=pd.cut(df['total_working_years'],bins=[0,2,5,10,40],labels=['Fresher', 'Junior', 'Mid', 'Senior'],include_lowest=True)
```
